```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline
```

```python
x = 6 * np.random.rand(200, 1) - 3
y = 0.8 * x**2 + 0.9 * x + 2 + np.random.rand(200, 1)
plt.plot(x,y,'g.')
# plt.show()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2,random_state=2)
model = LinearRegression()
model.fit(xtrain,ytrain)
```

```python
y_pred=model.predict(xtest)
r2score= r2_score(ytest,y_pred)
print(r2score)
```

```python
plt.plot(xtrain,model.predict(xtrain),'g')
plt.plot(x,y,'b.')
# plt.show()
```

```python
poly=PolynomialFeatures(degree=2)
x_train_trans=poly.fit_transform(xtrain)
x_test_trans=poly.transform(xtest)
print(xtrain[0])
print(x_train_trans[0])
```

```python
polymodel=LinearRegression()
polymodel.fit(x_train_trans,ytrain)
pred=polymodel.predict(x_test_trans)
print(r2_score(ytest,pred))
```

```python
print(polymodel.coef_)
print(polymodel.intercept_)
```
